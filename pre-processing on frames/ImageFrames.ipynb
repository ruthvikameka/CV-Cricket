```python
import cv2 
import os 
```

```python
cam = cv2.VideoCapture("D:\\Youtube\\Untitled.MOV") 
  
try: 
      
    
    if not os.path.exists('FRAMES'): 
        os.makedirs('FRAMES') 
  

except OSError: 
    print ('Error: Creating directory of data') 

```

```python
currentframe = 0
  
while(True): 
      
    
    ret,frame = cam.read() 
  
    if ret: 
        
        name = './FRAMES/frame' + str(currentframe) + '.jpg'
        print ('Creating...' + name) 
  
        
        cv2.imwrite(name, frame) 
  
       
        currentframe += 1
    else: 
        break
```

```python
cam.release() 
cv2.destroyAllWindows() 
```
